```python
##replace na values by grouping os, device brand
phone.reset_index(drop=True, inplace=True)
columns_to_process = ['rear_camera_mp', 'front_camera_mp', 'internal_memory', 'ram', 'battery', 'weight']

# Function to replace NA values with the median of the group
def replace_missing_with_median(x):
    return x.fillna(x.median())

# Apply the function within each group for each specified column
for column in columns_to_process:
    phone[column] = phone.groupby(['device_brand', 'os'])[column].transform(replace_missing_with_median)
```

```python
phone = pd.get_dummies(phone, columns=['device_brand','os','4g','5g'])
```

```python
y = pd.DataFrame(phone.normalized_used_price)
y.head(4)
type(y)
y.head()
```
